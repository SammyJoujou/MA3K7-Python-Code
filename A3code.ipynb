```python
import numpy as np
import seaborn as sns  #import all libraries
import matplotlib.pyplot as plt
from scipy import stats

def hatgame(): # Define the setup of the game
    numbers = np.arange(1, 200) #Array of numbers we're dealing with
    while len(numbers) > 1: # While loop to play the game until we get down to the final piece of paper
        x, y = np.random.choice(numbers, size=2, replace=False)  # Using numpys random number generator
        numbers = np.delete(numbers, np.where(numbers == x)) # I've never used the delete function before so I had some help from statology to do this
        numbers = np.delete(numbers, np.where(numbers == y))
        numbers = np.append(numbers, abs(x - y)) #Append the absolute value of the number to the list after getting rid of the old 
    return numbers[0]

def resultpool(iterations):  # Now define a new function to store all the other numbers
    results = []  # Empty results to start with
    for _ in range(iterations): # Repeat for as many iterations, blank place holder as we don't use it in the arguement.
        results.append(hatgame()) 
    return results

iterations = 1000 #Total number of iterations and store all the results in the total
total = resultpool(iterations)
    
sns.histplot(total, bins=50, color='red', edgecolor='black',  kde=True) # Use seaborn to plot all the results and a density line
plt.title('Hat Game Histogram')
plt.xlabel('Final Number') # Standard Labels
plt.ylabel('Frequency')
plt.show()
```

```python
def hatgame(hatsize): # Define the setup of the game
    numbers = np.arange(1, hatsize + 1) #Array of numbers we're dealing with
    while len(numbers) > 1: # While loop to play the game until we get down to the final piece of paper
        x, y = np.random.choice(numbers, size=2, replace=False)  # Using numpys random number generator
        numbers = np.delete(numbers, np.where(numbers == x)) # I've never used the delete function before so I had some help from statology to do this
        numbers = np.delete(numbers, np.where(numbers == y))
        numbers = np.append(numbers, abs(x - y)) #Append the absolute value of the number to the list after getting rid of the old 
    return numbers[0]

def resultpool(iterations, hatsizes):  #Now similiar to above example we define a function to store the numbers but this time for the 4 cases
    all_results = []  
    for size in hatsizes:  #We iterate through each of the 4 cases now
        results = []
        for _ in range(iterations):  # Appened the results to individual hat case but also to the total results
            results.append(hatgame(size))
        all_results.append(results)
    return all_results

iterations = 1000  #Total number of iterations
hatsizes = [50, 100, 250, 1000]   #Varying hat size numbers
all_results = resultpool(iterations, hatsizes)

fig, axs = plt.subplots(2, 2, figsize=(12, 8)) #Subplots in a 2x2 array 
axs = axs.flatten()
    
for i, ax in enumerate(axs):
    sns.histplot(all_results[i], bins=20, color='skyblue', edgecolor='black', kde=False, ax=ax) #Same histogram plot as before but iterating through i
    ax.set_title(f'Hat Game of size {hatsizes[i]}')
    ax.set_xlabel('Final Number')    #Normal axis conditions
    ax.set_ylabel('Frequency')
    mean = np.mean(all_results[i])
    median = np.median(all_results[i])  #Compute the mean and median and then plot a vertical line at these points
    ax.axvline(mean, color='red', linestyle='--', label=f'Mean: {mean:.2f}')
    ax.axvline(median, color='blue', linestyle='--', label=f'Median: {median:}')
    ax.legend()
    
plt.tight_layout()
plt.show()
```
